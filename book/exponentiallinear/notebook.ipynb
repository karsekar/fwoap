# Exponential versus linear growth

The choices are summarized again as follows:

* Choice 1: 100 dollars the first year, 300 dollars the second, 500 dollars the third, and so forth (for 25 total years)
* Choice 2: 1 penny the first year, 2 pennies the second, 4 the third, and so forth (for 25 total years)

Define and plot the trends below

```python
import numpy as np
import matplotlib.pyplot as plt

#define our time range
time = np.arange(25)

#compute our arrays
choice1 = (lambda x: 100+x*200)(time)
choice2 = (lambda x: 0.01*np.power(2,x))(time)

#make the figures
fig, (ax1, ax2) = plt.subplots(1,2,figsize=(8,4))

#make the linear scale plot first
ax1.plot(time,choice1,label="Choice 1")
ax1.plot(time,choice2,label="Choice 2")
ax1.legend()
ax1.set_xlabel('Time (years)')
ax1.set_ylabel('Pay out each year (dollars)')

#make the log scale for the second
ax2.set_yscale('log')
ax2.plot(time,choice1,label="Choice 1")
ax2.plot(time,choice2,label="Choice 2")
ax2.legend()
ax2.set_xlabel('Time (years)')
ax2.set_ylabel('Pay out each year (dollars; log scaled)')

#save the fig
fig.savefig('linearexponential.svg');
```
